```python
import random
random.seed(2026)
N = 10000
```

```python
moneda = ["Cara", "Cruz"]
primeras_tres_cara = 0
cuatro_caras_totales = 0

for N in range(N):
    lanzamiento1 = random.choice(moneda)
    lanzamiento2 = random.choice(moneda)
    lanzamiento3 = random.choice(moneda)
    lanzamiento4 = random.choice(moneda)
    lanzamiento5 = random.choice(moneda)
    lanzamiento6 = random.choice(moneda)

    lanzamientos1 = [lanzamiento1, lanzamiento2, lanzamiento3]
    lanzamientos2 = [lanzamiento4, lanzamiento5, lanzamiento6]
    if lanzamientos1.count("Cara") == 3:
        primeras_tres_cara += 1
        if lanzamientos2.count("Cara") == 1:
           cuatro_caras_totales += 1

p_total_4_caras_si_primeras_3_cara = cuatro_caras_totales/primeras_tres_cara

print(f"La probabilidad de que el total de caras sea 4 si los primeros 3 lanzamientos son cara es de {p_total_4_caras_si_primeras_3_cara:.4f}")
```

```python
caja = ["Roja"] * 5 + ["Azul"] * 3 + ["Verde"] * 2
suma_X = 0

for N in range(N):
    extraccion = random.sample(caja, 2)
    if extraccion[0] != extraccion[1]:
        X = 2
    else:
        X = 1
    suma_X += X
    
E_X = suma_X / N
print(f"El valor esperado de X es {E_X:.4f}")
```

```python
casos_segunda_roja = 0
casos_segunda_roja_y_primera_azul = 0

for N in range(N):
    caja = ["Roja"] * 5 + ["Azul"] * 3 + ["Verde"] * 2
    primera = random.choice(caja)
    caja += [primera]
    segunda = random.choice(caja)
    if segunda == "Roja":
        casos_segunda_roja += 1
        if primera == "Azul":
            casos_segunda_roja_y_primera_azul += 1
            
probabilidad = casos_segunda_roja_y_primera_azul / casos_segunda_roja
print(f"La probabilidad estimada es {probabilidad:.4f}")
```

```python

moneda = ["Cara", "Cruz"]
cinco_caras_o_mas = 0

for N in range(N):
    lanzamiento1 = random.choice(moneda)
    lanzamiento2 = random.choice(moneda)
    lanzamiento3 = random.choice(moneda)
    lanzamiento4 = random.choice(moneda)
    lanzamiento5 = random.choice(moneda)
    lanzamiento6 = random.choice(moneda)
    lanzamiento7 = random.choice(moneda)
    lanzamiento8 = random.choice(moneda)

    lanzamientos = [lanzamiento1, lanzamiento2, lanzamiento3, lanzamiento4, lanzamiento5, lanzamiento6, lanzamiento7, lanzamiento8]
    numero_de_caras = lanzamientos.count("Cara")

    if numero_de_caras >= 5:
        cinco_caras_o_mas += 1


p_cinco_o_mas_caras = cinco_caras_o_mas / N

print(f"La probabilidad de obtener cinco caras o más es de {p_cinco_o_mas_caras:.4f}")
```

```python
moneda = ["Cara", "Cruz"]
al_menos_una = 0
primera_cruz = 0

for N in range(N):
    lanzamiento1 = random.choice(moneda)
    lanzamiento2 = random.choice(moneda)
    lanzamiento3 = random.choice(moneda)
    lanzamiento4 = random.choice(moneda)

    lanzamientos = [lanzamiento1, lanzamiento2, lanzamiento3, lanzamiento4]
    numero_de_caras = lanzamientos.count("Cara")

    if numero_de_caras >= 1:
        al_menos_una += 1
        if lanzamientos[0] == moneda[1]:
            primera_cruz += 1

p_al_menos_una_cara_y_primera_cruz = primera_cruz / al_menos_una

print(f"La probabilidad de que la primera moneda haya sido cruz dado que al menos una fue cara es de {p_al_menos_una_cara_y_primera_cruz:.4f}")


```
