```python
import numpy as np 
import matplotlib.pyplot as plt
np.random.seed(2026)

N = 10000
```

```python
#Problema 1
dados = np.random.randint(1, 7, size=(N, 2))

dado1 = dados[:, 0]
dado2 = dados[:, 1]

suma_dados = dado1 + dado2

#a) Probabilidade de obter uma suma igual a 7
psuma= np.mean(suma_dados == 7)

#b) Probabilidade de obter uma suma igual a 7 dado que uno de los dados es par.
dados_par = (dado1 % 2 == 0) | (dado2 % 2 == 0)
psuma_dado_par = np.mean(suma_dados[dados_par] == 7)

print(f"Probabilidad de obtener una suma igual a 7: {psuma:.4f}")
print(f"Probabilidad de obtener una suma igual a 7 dado que uno de los dados es par: {psuma_dado_par:.4f}")
                         

```

```python
#Problema 2
monedas = np.random.choice(['cara', 'sello'], size=(N, 3))
moneda1 = monedas[:, 0]
moneda2 = monedas[:, 1]
moneda3 = monedas[:, 2]

#a) Probabilidad de obtener exactamente dos caras
caras = (moneda1 == 'cara') + (moneda2 == 'cara') + (moneda3 == 'cara') 
psuma_caras = np.mean(caras == 2)

#B) estimar E(x)
x = (moneda1 == 'cara') + (moneda2 == 'cara') + (moneda3 == 'cara')
ex = np.mean(x) 

print(f"Probabilidad de obtener exactamente dos caras: {psuma_caras:.4f}")
print(f"Valor esperado E(x): {ex:.4f}")
```
