# Ejercicio 2: Simulacion con monedas

Se simula el lanzamiento de tres monedas justas. Se estima la probabilidad de obtener exactamente dos caras y tambien el valor esperado de X, donde X es el numero de caras obtenidas.

```python
import random

# Fijamos la semilla para que los resultados se puedan repetir.
random.seed(2026)

# Numero exacto de repeticiones pedido.
repeticiones = 10000

# Lista con los dos posibles resultados de una moneda justa.
moneda = ["Cara", "Cruz"]

# Contador de experimentos donde salen exactamente dos caras.
casos_exactamente_dos_caras = 0

# Suma acumulada de todas las caras obtenidas en todos los experimentos.
suma_total_de_caras = 0

# Repetimos el experimento de lanzar tres monedas.
for repeticion in range(repeticiones):
    # Lanzamos tres monedas de forma independiente.
    lanzamiento1 = random.choice(moneda)
    lanzamiento2 = random.choice(moneda)
    lanzamiento3 = random.choice(moneda)
    
    # Guardamos los tres lanzamientos en una lista para contarlos facilmente.
    lanzamientos = [lanzamiento1, lanzamiento2, lanzamiento3]
    
    # Contamos cuantas caras aparecieron en esta repeticion.
    numero_de_caras = lanzamientos.count("Cara")
    
    # Si hay exactamente dos caras, aumentamos el contador de la parte a.
    if numero_de_caras == 2:
        casos_exactamente_dos_caras += 1
    
    # Sumamos siempre el numero de caras para estimar el valor esperado.
    suma_total_de_caras += numero_de_caras

# Parte a: probabilidad estimada de obtener exactamente dos caras.
probabilidad_dos_caras = casos_exactamente_dos_caras / repeticiones

# Parte b: promedio de caras por experimento, que estima E(X).
valor_esperado_x = suma_total_de_caras / repeticiones

# Imprimimos todas las respuestas numericas con exactamente 4 decimales.
print(f"Parte a - P(exactamente 2 caras): {probabilidad_dos_caras:.4f}")
print(f"Parte b - E(X): {valor_esperado_x:.4f}")
```

El valor esperado se estima promediando el numero de caras observado en las 10000 repeticiones.
