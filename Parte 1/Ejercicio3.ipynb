# Ejercicio 3: Canicas y Bayes

Se simulan extracciones de canicas sin reemplazo. En la parte a se estima la probabilidad de obtener dos canicas rojas. En la parte b se estima una probabilidad tipo Bayes: que la caja elegida haya sido la caja 1 dado que salio una canica roja y una verde.

```python
import random

# Fijamos la semilla antes de generar numeros aleatorios.
random.seed(2026)

# Numero exacto de repeticiones pedido.
repeticiones = 10000

# -------------------------
# Parte a
# -------------------------

# Representamos la caja con 5 rojas, 3 azules y 2 verdes.
caja = ["Roja"] * 5 + ["Azul"] * 3 + ["Verde"] * 2

# Contador de casos donde las dos canicas extraidas son rojas.
casos_dos_rojas = 0

# Repetimos la extraccion de dos canicas sin reemplazo.
for repeticion in range(repeticiones):
    # Extraemos dos canicas sin reemplazo usando random.sample.
    extraccion = random.sample(caja, 2)
    
    # Revisamos si ambas canicas extraidas son rojas.
    if extraccion[0] == "Roja" and extraccion[1] == "Roja":
        casos_dos_rojas += 1

# Calculamos la probabilidad estimada de obtener dos rojas.
probabilidad_dos_rojas = casos_dos_rojas / repeticiones

# -------------------------
# Parte b
# -------------------------

# Representamos las dos cajas del problema.
caja1 = ["Roja"] * 5 + ["Azul"] * 4 + ["Verde"] * 2
caja2 = ["Roja"] * 2 + ["Azul"] * 5 + ["Verde"] * 3

# Contador de casos donde ocurre el evento: una roja y una verde.
casos_evento = 0

# Contador de casos donde ocurre el evento y la caja elegida fue la caja 1.
casos_evento_y_caja1 = 0

# Repetimos el experimento de elegir caja y extraer dos canicas.
for repeticion in range(repeticiones):
    # Elegimos al azar la caja 1 o la caja 2 con igual probabilidad.
    numero_caja = random.choice([1, 2])
    
    # Segun la caja elegida, usamos la lista correspondiente.
    if numero_caja == 1:
        caja_elegida = caja1
    else:
        caja_elegida = caja2
    
    # Extraemos dos canicas sin reemplazo de la caja elegida.
    extraccion = random.sample(caja_elegida, 2)
    
    # El evento ocurre si hay exactamente una roja y una verde.
    evento_roja_y_verde = extraccion.count("Roja") == 1 and extraccion.count("Verde") == 1
    
    # Si ocurre el evento, aumentamos el contador del denominador.
    if evento_roja_y_verde:
        casos_evento += 1
        
        # Si ademas la caja fue la caja 1, aumentamos el numerador.
        if numero_caja == 1:
            casos_evento_y_caja1 += 1

# Calculamos la probabilidad de que fuera caja 1 dado que ocurrio el evento.
probabilidad_caja1_dado_evento = casos_evento_y_caja1 / casos_evento

# Imprimimos todas las respuestas numericas con exactamente 4 decimales.
print(f"Parte a - P(dos rojas): {probabilidad_dos_rojas:.4f}")
print(f"Parte b - P(caja 1 | una roja y una verde): {probabilidad_caja1_dado_evento:.4f}")
```

La parte b estima la probabilidad condicional contando solo las repeticiones donde el evento observado realmente ocurrio.
