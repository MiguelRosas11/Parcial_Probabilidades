# Ejercicio 1: Simulacion con dados

Se simula el lanzamiento de dos dados justos de seis caras. En la parte a se estima la probabilidad de que la suma sea 7. En la parte b se estima la probabilidad condicional de que la suma sea 7 dado que al menos uno de los dados es par.

```python
import random

# Fijamos la semilla para que la simulacion sea reproducible.
random.seed(2026)

# Numero exacto de repeticiones pedido para la simulacion.
repeticiones = 10000

# Contador de casos donde la suma de los dos dados es 7.
casos_suma_7 = 0

# Contador de casos donde al menos uno de los dados es par.
casos_al_menos_uno_par = 0

# Contador de casos donde pasan ambas cosas: suma 7 y al menos un dado par.
casos_suma_7_y_al_menos_uno_par = 0

# Repetimos el experimento muchas veces para estimar las probabilidades.
for repeticion in range(repeticiones):
    # Lanzamos dos dados justos usando numeros enteros del 1 al 6.
    dado1 = random.randint(1, 6)
    dado2 = random.randint(1, 6)
    
    # Calculamos la suma de los dos dados.
    suma = dado1 + dado2
    
    # Revisamos si la suma es igual a 7.
    suma_es_7 = suma == 7
    
    # Revisamos si al menos uno de los dados es par.
    al_menos_uno_par = (dado1 % 2 == 0) or (dado2 % 2 == 0)
    
    # Si la suma es 7, aumentamos el contador de la parte a.
    if suma_es_7:
        casos_suma_7 += 1
    
    # Si al menos un dado es par, contamos el caso condicionado.
    if al_menos_uno_par:
        casos_al_menos_uno_par += 1
    
    # Si pasan ambas condiciones, contamos el numerador de la probabilidad condicional.
    if suma_es_7 and al_menos_uno_par:
        casos_suma_7_y_al_menos_uno_par += 1

# Parte a: probabilidad estimada de que la suma sea 7.
probabilidad_suma_7 = casos_suma_7 / repeticiones

# Parte b: probabilidad estimada de suma 7 dado que al menos un dado es par.
probabilidad_condicional = casos_suma_7_y_al_menos_uno_par / casos_al_menos_uno_par

# Imprimimos todas las respuestas numericas con exactamente 4 decimales.
print(f"Parte a - P(suma = 7): {probabilidad_suma_7:.4f}")
print(f"Parte b - P(suma = 7 | al menos un dado par): {probabilidad_condicional:.4f}")
```

La parte b filtra los lanzamientos despues de simular dos dados normales, para no cambiar el experimento original.
