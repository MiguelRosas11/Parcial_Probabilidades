# Ejercicio 4: Simulacion con cartas

Se simula la extraccion de dos cartas sin reemplazo de una baraja estandar de 52 cartas. Se estima la probabilidad de obtener dos ases y se comparan P(A y B) con P(A)P(B) para decidir si los eventos A y B son independientes.

```python
import random

# Fijamos la semilla antes de cualquier generacion aleatoria.
random.seed(2026)

# Numero exacto de repeticiones pedido.
repeticiones = 10000

# Representamos la baraja con 4 ases y 48 cartas que no son ases.
baraja = ["As"] * 4 + ["No As"] * 48

# Contador de casos donde la primera carta es un as.
casos_a = 0

# Contador de casos donde la segunda carta es un as.
casos_b = 0

# Contador de casos donde ambas cartas son ases.
casos_a_y_b = 0

# Repetimos la extraccion de dos cartas sin reemplazo.
for repeticion in range(repeticiones):
    # Extraemos dos cartas sin reemplazo usando random.sample.
    extraccion = random.sample(baraja, 2)
    
    # La primera carta extraida representa el evento A.
    primera_carta = extraccion[0]
    
    # La segunda carta extraida representa el evento B.
    segunda_carta = extraccion[1]
    
    # Revisamos si ocurre A: la primera carta es un as.
    ocurre_a = primera_carta == "As"
    
    # Revisamos si ocurre B: la segunda carta es un as.
    ocurre_b = segunda_carta == "As"
    
    # Si ocurre A, aumentamos su contador.
    if ocurre_a:
        casos_a += 1
    
    # Si ocurre B, aumentamos su contador.
    if ocurre_b:
        casos_b += 1
    
    # Si ocurren A y B al mismo tiempo, aumentamos el contador conjunto.
    if ocurre_a and ocurre_b:
        casos_a_y_b += 1

# Parte a: probabilidad estimada de que ambas cartas sean ases.
probabilidad_dos_ases = casos_a_y_b / repeticiones

# Parte b: estimamos P(A), P(B), P(A y B), y P(A) * P(B).
probabilidad_a = casos_a / repeticiones
probabilidad_b = casos_b / repeticiones
probabilidad_a_y_b = casos_a_y_b / repeticiones
producto_probabilidades = probabilidad_a * probabilidad_b

# Imprimimos todas las respuestas numericas con exactamente 4 decimales.
print(f"Parte a - P(dos ases): {probabilidad_dos_ases:.4f}")
print(f"Parte b - P(A): {probabilidad_a:.4f}")
print(f"Parte b - P(B): {probabilidad_b:.4f}")
print(f"Parte b - P(A y B): {probabilidad_a_y_b:.4f}")
print(f"Parte b - P(A) * P(B): {producto_probabilidades:.4f}")

# Conclusion corta sobre independencia.
print("Conclusion: A y B no son independientes, porque al sacar la primera carta sin reemplazo cambia la probabilidad de que la segunda carta sea un as.")
```

Para eventos independientes deberia cumplirse P(A y B) = P(A)P(B). Aqui no se cumple porque las cartas se extraen sin reemplazo.
